```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
import math
plt.rcParams.update({'font.size': 15})
plt.rcParams['figure.figsize']= [10, 7]
```

# ALL graphs produced below may look different than in the report because they were used with particular and varying values of the initial conditions and parameters k, m

# Model 1

## Potential

U(x) = 1\2 kx^2 (1 - 2\3 ax)

## EoM
 m x" = -kx(1- ax)
 

```python
# we let:
a = 0.03
m = 20
k = 30
t = np.linspace(0, 10, 1001)
p = 4
cond_init = [10, 0]       #[x(t0), xd(t_0)]
omega_0 = np.sqrt(k/m)
```

```python
def model1(cond_init,t , m, k, a):
    
    x, xd = cond_init
    sys = [xd, (-k/m) * x * (1 - a * x)]
    
    return sys

```

```python
sol_model1 = odeint(model1, cond_init, t, args = (m, k, a))
```

```python
plt.plot(t, sol_model1[:,0], 'b', label='x(t)')
plt.plot(t, sol_model1[:, 1], 'g', label="xd(t)")
plt.legend(loc='upper center', bbox_to_anchor=(1.1, 1),  shadow=True, ncol=1)
plt.xlabel('t')
plt.grid()
plt.show()
```

## Nonharmonic effects

Nonharmonic effects induced by the a factor will lead to the period of oscillations changing depending on the amplitude of the oscillation. We observe this cheeky behaviour by plotting multiple amplitudes together.

```python
amplitudes = [3.3, 13.3, 23.3, 33.3]
x_vals = np.zeros((len(t),len(amplitudes)))
xdot_vals = np.zeros((len(t), len(amplitudes)))


for i in range(len(amplitudes)):
    init_vals = [amplitudes[i], 0]
    solution = odeint(model1, init_vals, t, args = (m, k, a))
    x_vals[:, i] = solution[:, 0]
    xdot_vals[:, i] = solution[:, 1]
```

```python
for i in range(len(amplitudes)):
    plt.plot(t, xdot_vals[:, i], label='xd(0)={}'.format(amplitudes[i]))
    plt.xlabel('time')
    plt.ylabel('xd(t)')
    

plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1),  shadow=True, ncol=1)
plt.grid()
plt.show()
```

# Model 2

## Potential

U(x) = 1\p k x^p

## EoM

m x" = -k x^(p - 1)

```python
def model2(cond_init, t, m, k, a, p):
    x, xd = cond_init
    sys = [xd, (-k/m) * x ** (p - 1)]
    
    return sys
    
```

```python
sol_model2 = odeint(model2, cond_init, t, args = (m, k, a, p))
```

```python
amplitudes = [3.3, 6.6, 9.9 ,13.3]
x_vals = np.zeros((len(t),len(amplitudes)))
xdot_vals = np.zeros((len(t), len(amplitudes)))


for i in range(len(amplitudes)):
    init_vals = [amplitudes[i], 0]
    solution = odeint(model2, init_vals, t, args = (m, k, a, p))
    x_vals[:, i] = solution[:, 0]
    xdot_vals[:, i] = solution[:, 1]
```

```python
for i in range(len(amplitudes)):
    plt.plot(t, x_vals[:, i], label='xd(0)={}'.format(amplitudes[i]))
    plt.xlabel('time')
#     plt.ylabel('xd(t)')
    

plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1),  shadow=True, ncol=1)
plt.grid()
plt.show()
```

## Plotting potentials for various powers, p

```python
ps = [2, 4, 6, 8, 10]
X = np.linspace(-3, 3, 100)

for i in range(len(ps)):
    potential = (1/ps[i]) * k * X**ps[i]
    plt.plot(X, potential, label = 'p = {}'.format(ps[i]))
    plt.xlabel('displacement')
    plt.ylabel('V(x)')
    plt.ylim(0, 30)
    
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1),  shadow=True, ncol=1)
plt.grid()
plt.show()
```

## Displacements for various p


```python
xs = np.zeros((len(t),len(ps)))
xdots = np.zeros((len(t), len(ps)))


for i in range(len(ps)):
    solution = odeint(model2, cond_init, t, args = (m, k, a, ps[i]))
    xs[:, i] = solution[:, 0]
    xdots[:, i] = solution[:, 1]
    plt.plot(t, xs[:, i], label = 'p = {}'.format(ps[i]))
    plt.xlabel('time')
    plt.ylabel('displacement')
    
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1),  shadow=True, ncol=1)
plt.grid()
plt.show()
```

## The model that includes a sinusoidal external force

```python
def model2extF(cond_init, t, m, k, a, p, F_0, omega):
    x, xd = cond_init
    F_ext = F_0 * np.sin(omega*t)
    sys = [xd, (F_ext/m) + (-k/m) * x ** (p - 1)]
    
    return sys
```

```python
F_0mag = [3]
omega = 0.9 * omega_0

for i in range(len(F_0mag)):
    solution = odeint(model2extF, cond_init, t, args = (m, k, a, p, F_0mag[i], omega))
    xs[:, i] = solution[:, 0]
    xdots[:, i] = solution[:, 1]
    plt.plot(t, xs[:, i], label = 'F ext. = {}'.format(F_0mag[i]))
    plt.xlabel('time')
    plt.ylabel('displacement')
    
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1),  shadow=True, ncol=1)
plt.grid()
plt.show()  
```

```python
omega_values = [0.1 * omega_0, 0.9 * omega_0 ,omega_0, 1.1 * omega_0, 10 * omega_0]

for i in range(len(omega_values)):
    solution = odeint(model2extF, cond_init, t, args=(m, k, a, p, F_0mag[0], omega_values[i]))
    xs[:, i] = solution[:, 0]
    xdots[:, i] = solution[:, 1]
    plt.plot(t, xs[:, i], label = 'omega = {:.2f}'.format(omega_values[i]))
    plt.xlabel('time')
    plt.ylabel('displacement')
    
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1),  shadow=True, ncol=1)
plt.grid()
plt.show()  
```

## The model that includes a frictional component

```python
def model4Frict(cond_init, t, m, k, a, p, F_0, omega, b):
    x, xd = cond_init
    F_ext = F_0 * np.sin(omega*t)
    
    sys = [xd, (F_ext/m) + (-k/m) * x ** (p - 1) - (b/m) * xd]
    
    return sys
```

## used to plot displacement for varying omega values

```python
bs = [1, 10, 15, 20, 50]

for i in range(len(omega_values)):
    solution = odeint(model4Frict, cond_init, t, args=(m, k, a, p, F_0mag[0], omega_values[i], bs[4]))
    xs[:, i] = solution[:, 0]
#     xdots[:, i] = solution[:, 1]
    plt.plot(t, xs[:, i], label = 'omega = {:.2f}'.format(omega_values[i]))
    plt.xlabel('time')
    plt.ylabel('displacement')
    
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1),  shadow=True, ncol=1)
plt.grid()
plt.show()  
```

## used to plot displacement for varying b values


The three cases to be investigated are:
- omega_0 ^2 > beta ^2 (underdamping)
- omega_0 ^2 = beta ^2 (critical damping)
- omega_0 ^2 < beta ^2 (overdamping)

```python
b_crit = math.sqrt(4 * k * m)
b_underd = (b_crit) * 0.7
b_overd = (b_crit) * 1.3

bvalsdamping = [b_crit, b_underd, b_overd]

xvalsdamping = np.zeros((len(t), len(bvalsdamping)))

for i in range(len(bvalsdamping)):
    solution = odeint(model4Frict, cond_init, t, args=(m, k, a, p, F_0mag[0], omega_values[0], bvalsdamping[i]))
    xvalsdamping[:, i] = solution[:, 0]
#     xdots[:, i] = solution[:, 1]
    plt.plot(t, xvalsdamping[:, i], label = 'b = {:.2f}'.format(bvalsdamping[i]))
    plt.xlabel('time')
    plt.ylabel('displacement')
    
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1),  shadow=True, ncol=1)
plt.grid()
plt.show()  
```

## Used for graphing the relationship between omega and maximum amplitude achieved 

```python
omega_values = np.linspace(0.1, 5, 200) * omega_0
xs = np.zeros((len(t),len(omega_values)))
res = np.zeros(len(omega_values))
beu = bs[4]

for i in range(len(omega_values)):
    solution = odeint(model4Frict, cond_init, t, args = (m, k, a, p, F_0mag[0], omega_values[i], beu))
    xs[:, i] = solution[:, 0]
    maxx = np.max(xs[:, i])
    res[i] = maxx
    
plt.scatter(omega_values, res)
plt.xlabel('omega')
plt.ylabel('max')
plt.title("plot for b = {}".format(beu))

plt.grid()
plt.show()
```
